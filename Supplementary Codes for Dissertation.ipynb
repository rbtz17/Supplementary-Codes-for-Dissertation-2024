```python
def check_volume(C):
    [n,k,d,q] = [C.length(), C.dimension(), C.minimum_distance(),H.base_field().order()]
    V = 0
    for i in range(floor((d-1)/2)+1):
        V += binomial(n,i)*(q-1)^i
    print(f"V = {V}, Bound = {q^(n-k)}")
    return V == q^(n-k)
def check_self_dual(C):
    CM = C.generator_matrix()
    compose = CM * CM.transpose() # Generator matrix act on itself
    return compose == compose * 0 # Self dual if this matrix is 0
```

```python
H  = codes.HammingCode(GF(2),3)
HD = H.dual_code()
HE  = codes.ExtendedCode(H)
G2  = codes.GolayCode(GF(2),extended=False)
G2D = G2.dual_code()
G3  = codes.GolayCode(GF(3),extended=False)
G3D = G3.dual_code()
EG2 = codes.GolayCode(GF(2))
EG3 = codes.GolayCode(GF(3))
```

```python
LC = [H, HD, HE, G2, G2D, EG2, G3, G3D, EG3]
LM = []
LML = []
LW = []
LWL = []

for C in LC:
    LM.append(C.generator_matrix())
    LML.append(latex(LM[-1]))
    LW.append(C.weight_enumerator())
    LWL.append(latex(LW[-1]))
    
def print_codes(LC,LM,LML):
    for i in range(len(LC)):
        print(LC[i],"\n")
        print(LML[i],"\n")
    return

def print_enumerator(LC,LW,LWL):
    for i in range(len(LC)):
        print(LC[i],"\n")
        print(LWL[i],"\n")
    return
```
